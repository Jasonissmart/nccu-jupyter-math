```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Contour設計

```python
x = np.linspace(-2*np.pi, 2*np.pi)
y = np.linspace(0, 4*np.pi)
[X,Y] = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)


plt.contourf(X,Y,Z, cmap="tab10", alpha=0.7)
plt.colorbar()
plt.contour(X,Y,Z, cmap="tab20", linestyles = "solid")
```
