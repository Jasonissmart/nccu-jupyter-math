```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

我們來看怎麼畫個不連續的函數, 比方說

$$f(x) = \tan(x)$$

```python
x = np.linspace(-5, 5, 1000)
y = np.tan(x)
```

```python
plt.plot(x, y)
```

```python
y[y>10] = np.inf
y[y<-10] = -np.inf
```

```python
plt.plot(x, y)
```
