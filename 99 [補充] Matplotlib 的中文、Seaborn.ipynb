如果明明有的字型, `matplotlib` 說找不到的話, 有可能需要讓 `matplotlib` 清掉原本的 cache。

    !rm -rf {mpl.get_cachedir()}

記得, 這不是每要做的事, 只有在有新的字型加入後做就可以。

```python
mpl.rc('font', family='Noto Sans CJK TC')
```

使用 Google `Noto Sans CJK TC` 字型, 請到[這裡](https://www.google.com/get/noto/#sans-hant)下載。

```python
x = np.linspace(-2*np.pi, 2* np.pi, 200)
y = np.sin(x)
```

```python
plt.title('$\sin$ 函數的圖形')
plt.plot(x,y)
```

```python
import seaborn as sns
```

```python
sns.set(rc={'font.family':'Noto Sans CJK TC'})
```

一般來說, 你只要用

    sns.set()
    
之後 `plt.plot` 等就「交給 `seaborn`」管了! 於是你可以再下之前我們的指令!

    mpl.rc('font', family='Noto Sans CJK TC')

```python
plt.title('$\sin$ 函數的圖形')
plt.plot(x,y)
```

### 回復到原來的 `matplotlib`

```python
sns.reset_orig()
```

```python
plt.title('$\sin$ 函數的圖形')
plt.plot(x,y)
```
